Shopping Variablility in Lausanne, Ceresio, & Riva

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("combined_group_data_two_cleaned.csv")

retail_df = df[df["Data Type"].astype(str).str.contains("Retail", case=False, na=False)]
retail_counts = retail_df["Town"].value_counts()

fig, ax = plt.subplots(figsize=(8, 5))
retail_counts.plot(kind="bar", ax=ax)

ax.set_title("Retail Activity by Town")
ax.set_xlabel("Town")
ax.set_ylabel("Number of Retail Locations")
plt.xticks(rotation=0)
plt.tight_layout()

fig.savefig("retail_chart.png", dpi=300)
plt.show()
```
