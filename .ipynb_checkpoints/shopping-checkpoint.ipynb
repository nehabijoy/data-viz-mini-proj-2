Shopping Variablility in Lausanne, Ceresio, & Riva

```python
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------
# 1. Load your combined dataset
# -----------------------------
df = pd.read_csv("combined_group_data_two_cleaned.csv")

# -----------------------------
# 2. Filter only Retail
# -----------------------------
retail_df = df[df["Data Type"] == "Retail"]

# -----------------------------
# 3. Count per town
# -----------------------------
retail_counts = retail_df["Town"].value_counts()

# -----------------------------
# 4. Plot
# -----------------------------
retail_counts.plot(kind="bar")

plt.title("Retail Locations by Town")
plt.xlabel("Town")
plt.ylabel("Number of Retail Locations")
plt.xticks(rotation=45)

plt.show()
```
